```python
#Right Outer Join using Left Outer Join bcz it doesn't support direct command of Right outer Join
df = pd.read_sql_query("""
SELECT
    EmployeeDemo.EmployeeID,
    EmployeeDemo.FirstName,
    EmployeeDemo.LastName,
    EmployeeDemo.Age,
    EmployeeDemo.Gender,
    EmployeeSalary.Job,
    EmployeeSalary.EmployeeSalary
FROM EmployeeSalary
LEFT OUTER JOIN EmployeeDemo ON EmployeeSalary.EmployeeID = EmployeeDemo.EmployeeID""", conn)
display(df)
```

```python
#Below query perform a Full Outer Join from EmployeeDemo to EmployeeSalary and use of union is to unite Left Outer Join of EmployeeDemo to EmployeeSalary from Left Outer Join of EmployeeSalary to EmployeeDemo. After unino perform a Left Outer Join of EmployeeSalary to EmployeeDemo.
df = pd.read_sql_query("""
SELECT
EmployeeDemo.EmployeeID,
EmployeeDemo.FirstName,
EmployeeDemo.LastName,
EmployeeDemo.Age,
EmployeeDemo.Gender,
EmployeeSalary.Job,
EmployeeSalary.EmployeeSalary
FROM EmployeeDemo
LEFT JOIN EmployeeSalary ON EmployeeDemo.EmployeeID = EmployeeSalary.EmployeeID

UNION

SELECT
EmployeeDemo.EmployeeID, NULL, NULL, NULL, NULL,
EmployeeSalary.Job, EmployeeSalary.EmployeeSalary
FROM EmployeeSalary
LEFT JOIN EmployeeDemo ON EmployeeSalary.EmployeeID = EmployeeDemo.EmployeeID
WHERE EmployeeDemo.EmployeeID IS Null""", conn)
df
```

```python
df = pd.read_sql_query("""
SELECT
EmployeeDemo.EmployeeID,
EmployeeDemo.FirstName,
EmployeeDemo.LastName,
EmployeeDemo.Age,
EmployeeDemo.Gender,
EmployeeSalary.Job,
EmployeeSalary.EmployeeSalary
FROM EmployeeDemo
LEFT JOIN EmployeeSalary ON EmployeeDemo.EmployeeID = EmployeeSalary.EmployeeID

UNION ALL

SELECT
EmployeeDemo.EmployeeID, NULL, NULL, NULL, NULL,
EmployeeSalary.Job, EmployeeSalary.EmployeeSalary
FROM EmployeeSalary
LEFT JOIN EmployeeDemo ON EmployeeSalary.EmployeeID = EmployeeDemo.EmployeeID
WHERE EmployeeDemo.EmployeeID IS Null""", conn)
df
```

```python
#Case statements
df = pd.read_sql_query("""
SELECT FirstName, LastName, Age,
CASE
    WHEN Age > 30 THEN 'Old'
    ELSE 'Young'
END AS 'Age Group'
FROM EmployeeDemo
WHERE Age is NOT NULL
ORDER BY Age""", conn)
display(df)
```

```python
df = pd.read_sql_query("""
SELECT FirstName, LastName, Age,
CASE
    WHEN Age > 30 THEN 'Old'
    WHEN Age BETWEEN 27 AND 30 THEN 'Young'
    ELSE 'Child'
END AS 'Age Group'
FROM EmployeeDemo
WHERE Age is NOT NULL
ORDER BY Age""", conn)
display(df)
```

```python
df = pd.read_sql_query("""
SELECT FirstName, LastName, Age,
CASE
WHEN Age = 30 THEN 'Stanley'
WHEN Age > 30 THEN 'Old'
WHEN Age BETWEEN 27 AND 30 THEN 'Young'
ELSE 'Child'
END AS 'Age Group'
FROM EmployeeDemo
WHERE Age is NOT NULL
ORDER BY Age""", conn)
display(df)
```

```python
df = pd.read_sql_query("""
SELECT FirstName, LastName, Job, EmployeeSalary,
CASE
WHEN Job='Salesman' THEN EmployeeSalary + (EmployeeSalary * .10)
WHEN Job='Accountant' THEN EmployeeSalary + (EmployeeSalary * .05)
WHEN Job='HR' THEN EmployeeSalary + (EmployeeSalary * .000001)
ELSE EmployeeSalary + (EmployeeSalary * .03)
END  AS EmployeeSalaryAfterRaise
FROM EmployeeDemo
JOIN EmployeeSalary
ON EmployeeDemo.EmployeeID = EmployeeSalary.EmployeeID""", conn)
display(df)
```

```python
#Having Clause
df = pd.read_sql_query("""
 SELECT Job, COUNT(Job)
 FROM EmployeeDemo JOIN EmployeeSalary
 ON EmployeeDemo.EmployeeID = EmployeeSalary.EmployeeID
 GROUP BY Job
 HAVING COUNT(Job) > 1""",conn)
display(df)
```

```python
df = pd.read_sql_query("""
SELECT Job, AVG(EmployeeSalary)
FROM EmployeeDemo JOIN EmployeeSalary
ON EmployeeDemo.EmployeeID = EmployeeSalary.EmployeeID
GROUP BY Job
HAVING AVG(EmployeeSalary)>45000
ORDER BY AVG(EmployeeSalary)""" , conn)
display(df)
```

```python
df = pd.read_sql_query("""
SELECT Demo.EmployeeID, Sal.EmployeeSalary
FROM EmployeeDemo AS Demo
JOIN EmployeeSalary AS Sal
ON Demo.EmployeeID = Sal.EmployeeID""", conn)
display(df)
```

```python
#Partition by
df = pd.read_sql_query("""
SELECT FirstName, LastName, Gender, EmployeeSalary, COUNT(Gender)
OVER (PARTITION BY Gender) as TotalGender
FROM EmployeeDemo Demo
JOIN EmployeeSalary Sal
ON Demo.EmployeeID = Sal.EmployeeID""", conn)
display(df)
```

```python
df = pd.read_sql_query("""
SELECT Gender, COUNT(Gender)
FROM EmployeeDemo Demo JOIN EmployeeSalary Sal
ON Demo.EmployeeID = Sal.EmployeeID GROUP BY Gender""", conn)
display(df)
```
